```python
# In this step I will apply cryptographic hashing to PII info
from pyspark.sql.functions import sha2, concat_ws, lit

df_fin_tran_2025 = spark.table("fraud_bronze.fraud_transactions_2025")

df_fin_tran_2025 = df_fin_tran_2025.withColumnRenamed("transaction_id", "transaction_id_raw").withColumnRenamed("sender_account", "sender_account_raw").withColumnRenamed("receiver_account", "receiver_account_raw").withColumn("transaction_id", sha2(concat_ws("_", lit("fin2025"), "transaction_id_raw"), 256)).withColumn("sender_account_id", sha2(concat_ws("_", lit("fin2025"), "sender_account_raw"), 256)).withColumn("receiver_account_id", sha2(concat_ws("_", lit("fin2025"), "receiver_account_raw"), 256)).drop("transaction_id_raw", "sender_account_raw", "receiver_account_raw")

df_fin_tran_2025.write.format("delta").mode("overwrite").saveAsTable("fraud_silver.fraud_transactions_2025")
                                                                                                                                                                                                                                                                                                                                                                                                                                                            

```
