```python
# In this step I will apply cryptographic hashing to PII info

from pyspark.sql.functions import sha2, concat_ws, lit

df_paysim = spark.table("fraud_bronze.paysim_transactions")

df_paysim = df_paysim.withColumnRenamed("nameOrig", "account_orig_raw").withColumnRenamed("nameDest", "account_dest_raw").withColumn("account_orig_id", sha2(concat_ws("_", lit("paysim"), "account_orig_raw"), 256)).withColumn("account_dest_id", sha2(concat_ws("_", lit("paysim"), "account_dest_raw"), 256)).drop("account_orig_raw", "account_dest_raw")

df_paysim.write.format("delta").mode("overwrite").saveAsTable("fraud_silver.paysim_transactions")
```
