```python
transaction_features = spark.table("ml_layer.transaction_features")
application_features = spark.table("ml_layer.application_features")


# In this step, I will start handling missing values for categorical variables in transaction_features table


transaction_features = transaction_features.fillna({
    "merchant_category": "unknown",
    "device_type": "unknown",
    "channel": "unknown",
    "location_city": "unknown",
    "amount": 0,
    "account_balance": 0,
    "customer_avg_transaction_amount": 0
})


transaction_features.write.format("delta")\
    .mode("overwrite")\
    .saveAsTable("ml_layer.transaction_features_clean")


# Now the same process in application_features table

application_features = application_features.dropna(
    subset=["account_id", "is_fraud"]
)

application_features = application_features.fillna({
    "payment_type": "unknown",
    "address_stability": 0,
    "name_email_similarity": 0,
    "days_since_request": 0,
    "under_25": 0,
    "zip_count_4w": 0,
})

application_features.write.format("delta")\
    .mode("overwrite")\
    .saveAsTable("ml_layer.application_features_clean")
```
