```python
# Here I will add the timestamp and source of the dataset

from pyspark.sql.functions import current_timestamp, lit

df_fraud_det_2023 = spark.table("fraud_raw.fraud_detection_2023").withColumn("dataset_source", lit("fraud_detection_2023")).withColumn("ingestion_ts", current_timestamp())

# Here I are writing the data to the bronze table

df_fraud_det_2023.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("fraud_bronze.fraud_transactions_2023")
```
