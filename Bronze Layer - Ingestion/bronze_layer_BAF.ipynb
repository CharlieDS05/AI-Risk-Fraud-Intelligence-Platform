```python
# Here I will add the timestamp and source of the dataset

from pyspark.sql.functions import current_timestamp, lit

# BAF base ds

df_baf_base = spark.table("fraud_raw.baf_base").withColumn("dataset_source", lit("baf_base")).withColumn("ingestion_ts", current_timestamp())

# Here I are writing the data to the bronze table

df_baf_base.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("fraud_bronze.baf_base")


# BAF variant_3 ds

df_variant_3 = spark.table("fraud_raw.baf_variant_3").withColumn("dataset_source", lit("baf_variant_3")).withColumn("ingestion_ts", current_timestamp())

# Here I are writing the data to the bronze table

df_variant_3.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("fraud_bronze.baf_variant_3")


# BAF variant_5 ds

df_variant_5 = spark.table("fraud_raw.baf_variant_5").withColumn("dataset_source", lit("baf_variant_5")).withColumn("ingestion_ts", current_timestamp())

# Here I are writing the data to the bronze table

df_variant_5.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("fraud_bronze.baf_variant_5")
```
