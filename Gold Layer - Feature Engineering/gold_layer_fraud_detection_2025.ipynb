```python
from pyspark.sql.functions import col, lit

fraud_2025_gold = (
    spark.table("fraud_silver.fraud_transactions_2025")
    .select(
        col("sender_account_id").cast("string").alias("customer_id"),
        col("amount").cast("double"),
        col("transaction_type").cast("string"),
        col("merchant_category").cast("string"),
        col("is_fraud").cast("boolean").cast("bigint").alias("is_fraud"),
        col("dataset_source").cast("string"),
        col("location").cast("string").alias("location_city"),
        col("payment_channel").cast("string").alias("channel"),
        col("device_used").cast("string").alias("device_type"),
        col("timestamp").alias("transaction_timestamp"),
        lit(None).cast("double").alias("account_balance"),
        col("transaction_id").cast("string")        
    )
)

fraud_2025_gold.write.format("delta")\
    .mode("overwrite").option("overwriteSchema", "true")\
    .saveAsTable("fraud_gold.fraud_2025_standarized")
```
