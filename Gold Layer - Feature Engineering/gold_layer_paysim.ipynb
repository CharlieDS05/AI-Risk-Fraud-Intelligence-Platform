```python
# Here I will start standarization process with Spark, so data will be ready for modeling
# To start, Spark libraries/functions will be imported

from pyspark.sql.functions import col, lit, to_timestamp

paysim_gold = (
    spark.table("fraud_silver.paysim_transactions")
    .select(
        col("account_orig_id").cast("string").alias("customer_id"), 
        col("amount").cast("double"), 
        col("type").alias("transaction_type"), 
        col("isFraud").cast("LONG").alias("is_fraud"), 
        col("dataset_source"),
        col("oldbalanceOrg").cast("double").alias("account_balance"), 
        lit(None).cast("string").alias("location_city"), 
        lit(None).cast("string").alias("channel"), 
        lit(None).cast("timestamp").alias("transaction_timestamp"),
        lit(None).cast("string").alias("device_type"),
        lit(None).cast("string").alias("merchant_category"), 
        lit(None).cast("string").alias("transaction_id")
    )
)

paysim_gold.write.format("delta")\
    .mode("overwrite").option("mergeSchema", "true")\
    .saveAsTable("fraud_gold.paysim_standarized")
```
