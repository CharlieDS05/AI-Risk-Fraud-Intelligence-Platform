```python
from functools import reduce

paysim_gold = spark.table("fraud_gold.paysim_standarized")
fraud_2023_gold = spark.table("fraud_gold.fraud_2023_standarized")
fraud_2025_gold = spark.table("fraud_gold.fraud_2025_standarized")

gold_transactions = reduce(
    lambda df1, df2: df1.unionByName(df2, allowMissingColumns=True),
    [paysim_gold, fraud_2023_gold, fraud_2025_gold]
)

gold_transactions.write.format("delta")\
    .mode("overwrite")\
    .saveAsTable("fraud_unified.unified_fraud_transactions")
```
