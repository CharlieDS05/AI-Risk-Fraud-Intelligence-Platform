```python
# Here I will add the timestamp and source of the dataset

from pyspark.sql.functions import current_timestamp, lit

df_paysim = spark.table("fraud_raw.paysim").withColumn("dataset_source", lit("paysim_synthetic")).withColumn("ingestion_ts", current_timestamp())

df_paysim.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("fraud_bronze.paysim_transactions")
```
