```python
from pyspark.sql.functions import col, lit, to_timestamp

fraud_2023_gold = (
    spark.table("fraud_silver.fraud_transactions_2023")
    .select(
        col("CUSTOMER_ID").cast("string").alias("customer_id"),
        col("TX_AMOUNT").alias("amount"),
        col("TX_FRAUD").cast("LONG").alias("is_fraud"),
        col("dataset_source").cast("string"),        
        col("TX_DATETIME").cast("timestamp").alias("transaction_timestamp"),
        col("TRANSACTION_ID").cast("string").alias("transaction_id"),
        col("TERMINAL_ID").cast("string").alias("channel"),
        lit(None).cast("string").alias("transaction_type"),
        lit(None).cast("string").alias("location_city"),
        lit(None).cast("string").alias("merchant_category"),
        lit(None).cast("double").alias("account_balance"),
        lit(None).cast("string").alias("device_type")        
    )
)

fraud_2023_gold.write.format("delta")\
    .mode("overwrite").option("mergeSchema", "true")\
    .saveAsTable("fraud_gold.fraud_2023_standarized") 
```
