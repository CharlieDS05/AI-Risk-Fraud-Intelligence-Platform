```python
# In this step I will apply cryptographic hashing to PII info

from pyspark.sql.functions import sha2, concat_ws, lit

df_fin_tran_2023 = spark.table("fraud_bronze.fraud_transactions_2023")

df_fin_tran_2023 = df_fin_tran_2023.withColumnRenamed("TRANSACTION_ID", "transaction_id_raw_2023").withColumnRenamed("CUSTOMER_ID", "customer_id_raw_2023").withColumn("TRANSACTION_ID", sha2(concat_ws("_", lit("fin2023"), "transaction_id_raw_2023"), 256)).withColumn("CUSTOMER_ID", sha2(concat_ws("_", lit("fin2023"), "customer_id_raw_2023"), 256)).drop("transaction_id_raw_2023", "customer_id_raw_2023")


df_fin_tran_2023.write.format("delta").mode("overwrite").saveAsTable("fraud_silver.fraud_transactions_2023")
                                                                                                                                                                                                                                                                                                                                                                                                                                                            

```
